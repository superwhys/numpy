## 英国和美国各自youtube1000的数据结合之前的matplotlib绘制出各自的评论数量的直方图


```python
from matplotlib import pyplot as plt
from matplotlib import rc
import numpy as np
import math
```

```python
%matplotlib inline
```

```python
us_file_path = './youtube_video_data/US_video_data_numbers.csv'
uk_file_path = './youtube_video_data/GB_video_data_numbers.csv'
```

```python
t_us = np.loadtxt(us_file_path,delimiter=',',dtype='int')
```

```python
#取评论数据
t_us_comments = t_us[:,-1]
#选择比5000小的数据
t_us_comments = t_us_comments[t_us_comments <= 5000]
print(t_us_comments)
print(np.max(t_us_comments))  #最大值
```

```python
#组距
d = 250
bin_nums = math.ceil((np.max(t_us_comments)-np.min(t_us_comments)) / d)
```

```python
#绘制直方图
plt.figure(figsize = (20,10),dpi = 80)
plt.hist(t_us_comments,bin_nums)
plt.grid(alpha = 0.3)
plt.show()
```

# 希望了解英国的youtube中视频的评论数和喜欢数的关系(散点图)

```python
t_uk = np.loadtxt(uk_file_path,delimiter=',',dtype='int')
```

```python
#选择喜欢数比50w小的数据
t_uk = t_uk[t_uk[:,1]<=500000]

t_uk_comment = t_uk[:,-1]
t_uk_like = t_uk[:,1]
```

```python
#设置中文字体
font = {'family':'simsun',
        'weight':'bold'}
rc('font',**font)
```

```python
plt.figure(figsize = (20,10),dpi = 80)
plt.scatter(t_uk_like,t_uk_comment)
plt.xlabel('喜欢数:')
plt.ylabel('评论数:')
plt.title('英国的youtube中视频的评论数和喜欢数的关系')
plt.show()
```
